# <font color='teal'>**Data integration strategy**</font>

## **Schema Mapping**
- Input 1 (cleaned_international_students_info.csv):
- Columns: ['StudentID', 'Name', 'Country', 'CGPA', 'LevelOfStudy', ...]
- Data Types: [string, string, string, float, string, ...]
- Input 2 (cleaned_international_students_Form_Responses.csv):
- Columns: ['StudentID', 'LevelOfStudy', 'HomeRegion', 'StudyDurationUI',
'CampusParticipation', 'CGPA']
- Data Types: [string, string, string, int, string, float]
- Common Columns for Integration: ['StudentID', 'LevelOfStudy', 'CGPA', 'CampusParticipation']

```python
import pandas as pd

# Load the CSV files
file_path_1 = "cleaned_international_students_info.csv"
file_path_2 = "cleaned_international_students_Form_Responses.csv"

# Read the datasets into DataFrames
students_info_df = pd.read_csv(file_path_1)
form_responses_df = pd.read_csv(file_path_2)

# Standardize column names for easier merging
students_info_df.columns = students_info_df.columns.str.strip()
form_responses_df.columns = form_responses_df.columns.str.strip()

# Create a mapping for column renaming (if needed)
rename_mapping = {
    'What is your current level of study?': 'LevelOfStudy',
    'How long have you been studying at the University of Idaho?': 'StudyDurationUI',
    'Which region does your home country belong to?': 'HomeRegion',
    'What is your current Cumulative Grade Point Average (CGPA)?': 'CGPA',
    'How often do you participate in campus activities and events?': 'CampusParticipation',

}

# Rename columns in the form responses DataFrame
form_responses_df = form_responses_df.rename(columns=rename_mapping)

# Perform an inner join on the common columns
# Choose 'LevelOfStudy', 'HomeRegion', and 'CGPA' as the common columns
integrated_df = pd.merge(students_info_df, form_responses_df, on=['LevelOfStudy', 'HomeRegion', 'CGPA', 'CampusParticipation'], how='inner')


integrated_df.head()
```

```python
#   Check for missing values
missing_values = integrated_df.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

```python
integrated_df.dropna (axis = 0, how ='any')
```

```python
#   Check for duplicates and remove if necessary
duplicates = integrated_df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
integrated_df.drop_duplicates(inplace=True)
```

```python
# Save the integrated dataset
integrated_df.to_csv('international_students_integrated_dataset', index=False)
```
